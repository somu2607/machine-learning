```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("/content/Diwali Sales Data (2).csv", encoding ='unicode_escape')

```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop_duplicates()
dataset['Amount'] = dataset['Amount'].fillna(dataset['Amount'].mean())
dataset['Amount'].dtypes
```

```python
dataset.columns
```

```python
sales_gen = dataset.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.bar(sales_gen['Gender'], sales_gen['Amount'], color=['pink','violet'])
plt.title('Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.show()
```
